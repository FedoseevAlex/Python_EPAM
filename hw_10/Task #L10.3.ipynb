### Reading csv to df

```python
df = pd.read_csv('/home/linmachine/Documents/Python_EPAM/hw_10/Summer Olympic medallists 1896 to 2008 - ALL MEDALISTS.csv', skiprows=4) 
```

#### Число всех медалей, выйгранных на каждой олимпиаде.

```python
medal_count = df.groupby(['City', 'Edition']).Medal.agg('count')
```

```python
medal_count.plot(kind='bar', figsize=(10, 6))
```

#### Число всех медалей, выйгранных каждой страной за все время.

```python
country_medals = df.groupby(['NOC']).Edition.agg(['min', 'max', 'count']).reset_index(level=0)
country_medals
```
