### Imports:

```python
import pandas as pd
```

### Reading csv to df

```python
df = pd.read_csv('./Summer Olympic medallists 1896 to 2008 - ALL MEDALISTS.csv', skiprows=4) 
```

```python
pp = r'\w*\d{2,4}m'
```

```python
df[(df.Athlete == 'OWENS, Jesse') & (df.Event.str.contains(pp))]
```

#### На какой дистанции Jesse Owens выйграл медаль?

```python
df[(df.Athlete == 'OWENS, Jesse') & (df.Event.apply(p.match))]
```

#### Какая страна выйграла большинство медалей в бадминтоне среди мужчин?

```python
dat = df[(df.Gender == 'Men') & (df.Discipline == 'Badminton') & (df.Medal == 'Gold')]
```

```python
country = dat.NOC.value_counts().index[0]
```

```python
dat[(dat.NOC == country)].sort_values('Athlete')
```

####  Какие три страны выйграли большинство медалей с 1984 по 2008 годы?

```python
df[(df.Edition >= 1984) & (df.Edition <= 2008)].NOC.value_counts().head(3)
```

```python
pd.DataFrame(df[(df.Edition >= 1984) & (df.Edition <= 2008)].NOC.value_counts().head(3))
```

#### Золотые медалисты по бегу на 100 метров.

```python
dat = df[(df.Event == '100m') & (df.Gender == 'Men') & (df.Medal == 'Gold')]
```

```python
dat.drop(['Sport', 'Discipline', 'Gender', 'Event', 'Event_gender', 'Medal'], axis=1).sort_values(by='Edition')
```
