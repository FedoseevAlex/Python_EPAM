### Imports:

```python
import pandas as pd
```

### Reading csv to df

```python
df = pd.read_csv('./Summer Olympic medallists 1896 to 2008 - ALL MEDALISTS.csv', skiprows=4) 
```

### Таблица топовых атлетов США по общему количеству медалей. 

```python
# Get USA sportsmen only
usa_athletes = df[(df.NOC == 'USA')]
# Group data by year, name, discipline and medals, then unstack by medals and fill NaN with zeros
grouped_athletes = usa_athletes.groupby(['Edition',
                                         'Athlete',
                                         'Discipline',
                                         'Medal']).size().unstack('Medal', fill_value=0)
# Make data look nicer
grouped_athletes = grouped_athletes.reset_index().reindex(columns=['Edition',
                                                                 'Athlete',
                                                                 'Discipline',
                                                                 'Gold',
                                                                 'Silver',
                                                                 'Bronze'])
grouped_athletes.columns.name = '№'
# Add 'Total' column
grouped_athlets['Total'] = grouped_athlets.loc[:, 'Gold': 'Bronze'].sum(axis=1)
# Create resulting athlets rating 
result_top = pd.DataFrame()
for year in grouped_athlets.Edition.unique():
    # Make subframe by year
    year_subframe = grouped_athlets[grouped_athlets.Edition == year]
    # Get athlets with maximum medals 
    top_athlets_by_year = year_subframe[(year_subframe.Total == year_subframe.Total.max())]
    # Add top athlets by year to resulting data frame
    result_top = result_top.append(top_athlets_by_year)
result_top = result_top.reset_index().drop('index', axis=1)
```
